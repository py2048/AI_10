```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://thachln.github.io/datasets/TuyenVN.csv')
```

```python
df
```

```python
df1 = pd.DataFrame.copy(df[0:5])
df1['name'] = ['Văn Lâm', 'Duy Mạnh', 'Văn Hậu', "Văn Thanh", "Trọng Hoàng"]
```

```python
df1
```

```python
fig = plt.figure(dpi=150)
plt.title('Age histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.hist(df['age'], bins=5, color='#124122')
plt.show()
```

```python
ig = plt.figure(dpi=150)
plt.title('Age histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.hist(df['height'], bins=10, color='#124122')
plt.show()
```

```python
plt.figure(dpi=150)
plt.title('Age bar chart')
plt.xlabel('Name')
plt.ylabel('Age')
plt.bar(df1['name'], df1['age'])
plt.show()
```
