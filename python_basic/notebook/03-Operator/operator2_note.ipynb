### Bài tập

Viết chương trình nhập từ bàn phím hai số nguyên `x` và `y`. Sau đó,
kiểm tra giá trị của `x` có lớn hơn giá trị `y` hay không. Nếu có (`x`
lớn hơn `y`) thì in ra `x > y: True`, ngược lại in ra `x > y: False`.

Ví dụ nếu bạn nhập `x = 10, y = 5` thì màn hình sẽ hiển thị ra:

``` markup
x > y: True
```

Nếu bạn nhập `x = 2, y = 8` thì màn hình sẽ hiển thị ra:

``` markup
x > y: False
```

Ví dụ cụ thể với Python:

``` python
x = 10
y = 12

print('x > y is', x > y)
print('x < y is', x < y)
print('x == y is', x == y)
print('x != y is', x != y)
print('x >= y is', x >= y)
print('x <= y is', x <= y)
```

Kết quả khi chạy chương trình:

``` markup
x > y is False
x < y is True
x == y is False
x != y is True
x >= y is False
x <= y is True
```
