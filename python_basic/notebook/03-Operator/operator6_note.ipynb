### **Bài tập**

Cho 4 số nguyên `x, y, z` và `t` được nhập từ bàn phím. Bạn hãy viết
chương trình để kiểm tra 4 giá trị này có thoả mãn điều kiện `x > y`
và `z < t` hay không. In ra màn hình `"Result evaluation is True"` nếu 4
số thoả mãn điều kiện; nếu không, in ra `"Result evaluation is False"`.

Ví dụ nếu bạn nhập `x = 5, y = 6, z = 8, t = 3` thì màn hình sẽ hiển thị
ra:\`\`

``` markup
Result evaluation is False
```

Nếu bạn nhập `x = 10, y = 3, z = 7, t = 13` thì màn hình sẽ hiển thị ra:

``` markup
Result evaluation is True
```
